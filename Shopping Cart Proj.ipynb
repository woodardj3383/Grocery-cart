### Shopping Cart

```python
##### from IPython.display import clear_output

class Cart:
    _all = []
    #def __init__(self,items):
     #   self.items = item
        
    def set_items(self, item):
        
        return self._all.append(item)
 
            
    def remove_item(self, item_s):
        for s in self._all:
            if item_s in [i.name for i in self._all]:
                self._all.remove(s)
                break
                
            else:
                print(f'That item is not in your cart: {self._all}')
                
#    
    def show_items(self):
       
        total = 0
        print("*"* 40)
        if not self._all:
            print("Your cart is currently empty.")
        else:
            display_cart = []
            for x in self._all:
                if x.name not in [i['name']for i in display_cart]:
                    display_cart.append({
                        'name': x.name,
                        'price': x.price
                    })
                else:
                    for i in display_cart:
                        if i['name'] == x.name:
                            i['price'] += x.price
           
            for i in range(len(display_cart)):
                print(f"{i+1}: {display_cart[i]['name']} [{[i.name for i in self._all].count(display_cart[i]['name'])}] ${display_cart[i]['price']}")
            print("*"* 40)
        
        total += float(sum([p.price for p in self._all]))

        print(f"Total: ${total:,.2f}")  
    
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = float(price)
        

    
class ShoppingCart:
    
       
    @classmethod    
    def show_instructions(self):
        print("""Type 'add' to add an item to your cart.
Type 'remove' to remove an item from your cart.
Type 'clear' to remove all items from your cart.
Type 'show' to show your cart items.
Type 'quit' to exit program and show your cart.
""")
    
    def run():
       
        ShoppingCart.show_instructions()
        
        done = False
        while not done:
            Cart().show_items()
            ans = input('What would you like to do? ').lower()
            clear_output()
            if ans == 'add':
                item_name = input('What would you like to add? ').lower()
                item_price = float(input('How much does this item cost? '))
                item_ = Item(item_name, item_price)
                Cart().set_items(item_)
               
            elif ans == 'quit':
                #Cart().show_items()
                done = True
            elif ans == 'remove':
                Cart().show_items()  
                r_item = input('What yould you like to remove? ').lower()
                Cart().remove_item(r_item)
            elif ans == 'clear':
                ask = input('Are you shure you want to clear the car? Y/N').lower()
                if ask == 'y':
                    Cart()._all.clear()
                else:
                    pass
            #if ans == 'show':
        Cart().show_items()      
                
                
    
    
ShoppingCart.run()   
```
